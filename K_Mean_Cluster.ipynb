```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
heights = np.random.uniform(150,200,100)
```

```python
heights
```

```python
heights.shape
```

```python
heights_reshaped = heights.reshape(-1,1)
```

```python
heights_reshaped.shape
```

```python
model = KMeans(n_clusters=3, random_state=42) #Creating an instance of KMeans with 3 clusters
```

```python
model.fit(heights_reshaped)
```

```python
cents = model.cluster_centers_
cents
```

```python
labels = model.labels_
labels
```

```python
plt.scatter(heights, np.zeros_like(heights), c=labels, cmap='rainbow', marker='o', s=30)
```

What is Inertia? It is the measure of how tightly grouped the data points are within each cluster. It is the sum of the squared of the distances between ech data point and the centroid of cluster to which it belongs.

```python
#Calculating the wcss for different values of k
wcss = []
for k in range(1,10):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(data)
    wcss.append(model.inertia_) #wcss values
```

```python
#plot the elbow method graph
plt.plot(range(1,10), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster sum of squares(WCSS)')
plt.show()
```
