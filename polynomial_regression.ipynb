```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('/content/plasma_physics_data.xlsx')
```

```python
data.head()
```

```python
#Finding the mathematical relationship btw our data
```

```python
fig, ax=plt.subplots(2,1)

ax[0].scatter(data['Magnetic'],data['Ion'])

ax[1].scatter(data['Magnetic'], data['Electron_Temp'])
```

```python
fig, ax=plt.subplots(2,1,figsize=(10,15))

ax[0].scatter(data['Magnetic'],data['Ion'],color ='blue', label = 'Ion_density')
ax[0].set_title('Mag Field vs Ion Den')
ax[0].set_xlabel('Magnetic (tesla)')
ax[0].set_ylabel('Ion density')
ax[0].legend()

ax[1].scatter(data['Magnetic'], data['Electron_Temp'],color ='red', label = 'Electron_Temp')
ax[1].set_title('Mag Field vs Electron temp')
ax[1].set_xlabel('Magnetic (tesla)')
ax[1].set_ylabel('Electron temp')
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
x=data['Magnetic'].values.reshape(-1,1)#very important when using sklearn to reshape data to 2 dimensional
y=data['Ion'].values
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model= LinearRegression()
linear_model.fit(x,y)
```

```python
y_linear_pred= linear_model.predict(x)
```

```python
plt.scatter(x,y,color='blue',label='Actual Data')

plt.scatter(x,y_linear_pred,color='red',label='Predicted Data')
plt.title('Checking out linear model')
plt.xlabel('Mag')
plt.ylabel('Ion')
plt.legend
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
x_poly = poly.fit_transform(x)
```

```python
x_poly.shape
```

```python
model_2 =LinearRegression()
```

```python
model_2.fit(x_poly,y)
```

```python
y_poly_pred=model_2.predict(x_poly)
```

```python
plt.scatter(x,y,color='blue',label='Actual Data')

plt.scatter(x,y_poly_pred,color='green',label='Predicted Data')
plt.title('Checking out polynomial model model')
plt.xlabel('Mag')
plt.ylabel('Ion')
plt.legend
plt.show()
```

```python
z = data['Electron_Temp'].values
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
x_poly = poly.fit_transform(x)
```

```python
model_3 =LinearRegression()
```

```python
model_3.fit(x_poly,z)
```

```python
z_poly_pred=model_3.predict(x_poly)
```

```python
plt.scatter(x,z,color='blue',label='Actual Data')

plt.scatter(x,z_poly_pred,color='green',label='Predicted Data')
plt.title('Checking out polynomial model model')
plt.xlabel('Mag')
plt.ylabel('Ion')
plt.legend
plt.show()
```

```python
poly = PolynomialFeatures(degree=3)
```

```python
x_poly2 = poly.fit_transform(x)
```

```python
model_4 =LinearRegression()
```

```python
model_4.fit(x_poly2,z)
```

```python
z_poly_pred2=model_4.predict(x_poly2)
```

```python
plt.scatter(x,z,color='blue',label='Actual Data')

plt.scatter(x,z_poly_pred2,color='green',label='Predicted Data')
plt.title('Checking out polynomial model model')
plt.xlabel('Mag')
plt.ylabel('Ion')
plt.legend
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae = mean_absolute_error(z,z_poly_pred2)
mase = mean_squared_error(z,z_poly_pred2)
r2 =r2_score(z,z_poly_pred)
```
